```python
import numpy as np

time_decrease=10
pay_decrease=5000

time_increase=15
pay_increase=10000


# задаём функцию парето для произвольных значений стартового времени до работы/зарплаты
def pareto(initial_point):
    start_time = initial_point[0]
    start_pay = initial_point[1]
    
    def pareto_front(time):
        if (time > start_time):
            return start_pay + (time - start_time) / time_increase * pay_increase
        else:
            return start_pay + (time - start_time) / time_decrease * pay_decrease

    return pareto_front
```

```python
# Отрисуем график

import matplotlib.pyplot as plt

pareto_setup = pareto([60, 100000])

x = np.linspace ( start = 0.    # нижний лимит
                , stop = 200     # верхний лимит
                , num = 50      # генерируем 50 точек. Можно меньше, тк функция линейна на двух отрезках
                )
y = [a for a in map(pareto_setup, x)]

plt.xlabel('time')
plt.ylabel('pay')

plt.plot(x,y,'-gD',markevery=[15]) # рисуем график функции. Отмечаем точку изменения линейного коэффициента

plt.show()
```
